```python
def get_sds_for_case(path):
    sds = dict()
    for p in Path(path).glob('**/*.dcm'):
        p = str(p)
        dcm = pydicom.dcmread(p, stop_before_pixels=True)
        sds[dcm.SeriesDescription] = sds[dcm.SeriesDescription]+1 if dcm.SeriesDescription in sds.keys() else 1
    return sds
```

```python
CS_SAX = re.compile('CS_1090_retro_2\-shot_\\d{1,2}sl_acc6\.5')
```

```python
for c in all_cases:
    print('Case: ', c)
    sds = get_sds_for_case(os.path.join(bp, c))
    success = False
    for k in sds.keys():
        match = CS_SAX.match(k)
        if not match: continue
        else:
            print('      ', match.group(0))
            success = True
    if not success: print('                  NO CS FOUND!!')
```

```python
def add_LL_tag(store_path, dcm, tag='Lazy Luna: None'): # Lazy Luna: SAX CS
    try:    dcm[0x0b, 0x10].value = tag
    except: dcm.private_block(0x000b, tag, create=True)
    dcm.save_as(filename=store_path, write_like_original=False)
```

```python
for c in all_cases:
    for ip, p in enumerate(Path(os.path.join(bp, c)).glob('**/*.dcm')):
        p = str(p)
        dcm = pydicom.dcmread(p, stop_before_pixels=False)
        match = CS_SAX.match(dcm.SeriesDescription)
        #print(dcm.SeriesDescription, bool(match))
        if not match: add_LL_tag(p, dcm, tag='Lazy Luna: None')
        else:         add_LL_tag(p, dcm, tag='Lazy Luna: SAX CS')
    print('Done for', c)
```
